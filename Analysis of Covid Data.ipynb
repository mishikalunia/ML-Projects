```python
import pandas as pd
df = pd.DataFrame(covid.get_data())
print(df)
```

```python
df.columns
```

```python
df = df.drop(df.index[7])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
p1 = df.sort_values('total_cases',ascending=False).head(10)
```

```python
p1
```

```python
figure(figsize=(10,5),dpi=100)
x=p1['country']
y=p1['total_cases']
plt.plot(x,y,'go--',label='Total Cases')
plt.xticks(rotation=90)
plt.legend()
plt.title('Total no. of cases in top 10 Countries')
plt.show()
```

```python
figure(figsize=(10,5),dpi=100)
x=p1['country']
y=p1['total_cases']
plt.bar(x,y,label='Total Cases',color ="mediumturquoise")
plt.xticks(rotation=90)
plt.legend()
plt.title('Total no. of cases in top 10 Countries')
plt.show()
```

```python
figure(figsize=(10,5),dpi=100)
x=p1['country']
y=p1['total_cases']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(y, labels = x,autopct='%1.2f%%')
plt.title('Total no. of cases in top 10 Countries')
plt.show()
```
